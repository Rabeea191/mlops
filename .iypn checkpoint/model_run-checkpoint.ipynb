```python
def wavelet_packet_based_feature_extractor(sr,embedding,np_data,columns):
    access_level=1
    max_level=6
    node_data=[]
    for y in np_data:
        wp = pywt.WaveletPacket(data=y, wavelet='db4', mode='smooth',maxlevel=max_level) #create wavelet packet object
        node_data.append(wp['adadad'].data)

    feats=[embedding(x,sr) for x in node_data]

    feats_df=pd.DataFrame(feats,columns=columns)
    return feats_df
```

```python
'''
Feature extraction method
'''
# librosa.effects.preemphasis(x)
def feature_extractor(sr,embedding,np_data,columns):
    feats=[embedding(x,sr) for x in np_data]
    feats_df=pd.DataFrame(feats,columns=columns)
    return feats_df

# ['aaaaa', 'aaaad', 'aaadd', 'aaada', 'aadda', 'aaddd', 'aadad', 'aadaa', 'addaa', 'addad', 'adddd', 'addda', 
# 'adada', 'adadd', 'adaad', 'adaaa', 'ddaaa', 'ddaad', 'ddadd', 'ddada', 'dddda', 'ddddd', 'dddad', 'dddaa',
#  'dadaa', 'dadad', 'daddd', 'dadda', 'daada', 'daadd', 'daaad', 'daaaa']
```

```python
def dwt_feature_extractor(sr,embedding,np_data,columns):
    lpcs=[]
    feats=[embedding(x,sr) for x in np_data]
    for ca ,cd in feats:
        cA_lpc=wavelet_exp(ca,sr)
        cD_lpc=wavelet_exp(cd,sr)
        lpc=[*cA_lpc,*cD_lpc]
        lpcs.append(lpc)
    feats_df=pd.DataFrame(lpcs,columns=columns)
    return feats_df
        
```

```python
def b_split(df):
#     test_size=int((df.shape[0]/100)*30)
#     train_size=df.shape[0]-test_size
    
    covid=df[df['status']=='covid-19']
    healthy=df[df['status']=='healthy']
    
    cov_train=covid.iloc[int((covid.shape[0]/100)*30):,:]
    cov_test=covid.iloc[:int((covid.shape[0]/100)*30),:]
    
    hea_train=healthy.iloc[int((healthy.shape[0]/100)*30):,:]
    hea_test=healthy.iloc[:int((healthy.shape[0]/100)*30),:]
    
    cov_hea_train=pd.concat([cov_train,hea_train],axis=0)
    cov_hea_test=pd.concat([cov_test,hea_test],axis=0)
    
    xtrain=cov_hea_train.iloc[:,0:len(cov_hea_train.columns)-1]
    ytrain=cov_hea_train['status']
    
    xtest=cov_hea_test.iloc[:,0:len(cov_hea_test.columns)-1]
    ytest=cov_hea_test['status']
    
    xtrain,xtest,ytrain,ytest=xtrain.reset_index(drop=True),xtest.reset_index(drop=True),ytrain.reset_index(drop=True),ytest.reset_index(drop=True)
    
    print(xtrain.shape,ytrain.shape,xtest.shape,ytest.shape)
    
    return xtrain,xtest,ytrain,ytest 
# b_split(df)
```
