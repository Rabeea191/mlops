audio feature extraction

```python
def features(y,sr):
    als=[]
    alss=[]
    positive=[]

    als=[]
    rms = np.mean(librosa.feature.rms(y=y))
    spec_cent = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))
    spec_bw = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr))
    rolloff = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr))
    zcr = np.mean(librosa.feature.zero_crossing_rate(y))

    mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40).T,axis=0)
    melspec= np.mean(librosa.feature.melspectrogram(y=y, sr=sr, n_mels=40, fmax=8000).T,axis=0)
    chroma_stft = np.mean(librosa.feature.chroma_stft(y=y, sr=sr, n_chroma=36).T,axis=0)
    chroma_cq = np.mean(librosa.feature.chroma_cqt(y=y, sr=sr, n_chroma=36).T,axis=0)
    chroma_cens = np.mean(librosa.feature.chroma_cens(y=y, sr=sr, n_chroma=36).T,axis=0)
    contrast = np.mean(librosa.feature.spectral_contrast(y=y, sr=sr).T,axis=0)
    tonnetz = np.mean(librosa.feature.tonnetz(y=librosa.effects.harmonic(y), sr=sr,).T,axis=0)

    positive.append([rms])
    positive.append([spec_cent])
    positive.append([spec_bw])
    positive.append([rolloff])
    positive.append([zcr])

    positive.append(mfcc)
    positive.append(melspec)
    positive.append(chroma_stft)
    positive.append(chroma_cq)
    positive.append(chroma_cens)
    positive.append(contrast)
    positive.append(tonnetz)
    pos=np.array(positive)
   
    for i in pos:
        for x in i:
            als.append(x)
            
    alss.append(als)


    return alss
```

```python
f_name='/home/bigpenguin/dataset_notebook/project_covid/covid_models/silent_split_covid_dataset/new_dataset/positive/'
wav=f_name
for i in os.listdir(wav):
    y, sr = librosa.load(wav+i,res_type='kaiser_fast')
    y = librosa.util.normalize(y)
    fea=pd.DataFrame(features(y,sr),columns=col_names)
    print(i," -- ",rf.predict(fea)[0],rf.predict_proba(fea)[0])
    
    if rf.predict(fea)[0]=='Non-cough':
        os.remove(wav+i)

    
```
